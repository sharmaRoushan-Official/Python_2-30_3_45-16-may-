```python
# copy the list
# there are two types copy function
# 1. clone copy
# 2. Deep copy
# ------------------------------------

#1. clone copy[shallow copy]
list1=["Hello world"]
print(list1)
new_list=list1.copy()
print(new_list)

# updating the list
list1.append('Shrinivas')
print(list1)
```

```python
list1
```

```python
new_list
```

```python
# Deep copy

list3=["Python"]
print("Original List:",list3)
list4=list3
print("Duplicate List:",list4)


# Adding element in the duplicate list
list4.append('Modiji')
print(list4)
```
